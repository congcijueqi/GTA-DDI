```python
import csv

# 读取 drug_list_deepddi.csv 文件
with open('drug_list_deepddi.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # 跳过标题行
    data = list(reader)

# 生成第一列和第二列数据
output_data = []
for i, row in enumerate(data):
    drugbank_id = row[0]
    cid = f"CID{int(drugbank_id[2:]):08d}"
    output_data.append([i, cid] + row)

# 将生成的数据写入到 CSV 文件中
with open('drug_list_deepddi.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['index', 'CID', 'drugbank_id', 'smiles'])  # 添加新的列标题
    writer.writerows(output_data)

print("生成完成，已保存到 output.csv 文件中。")
```

```python
import csv

# 读取原始数据文件
with open('all_data_deepddi.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # 跳过标题行
    data = list(reader)

# 生成新的 drugbank_id 和 cid
drugbank_id_counter = 1
cid_counter = 1
drugbank_id_dict = {}
cid_dict = {}

output_data = []
for row in data:
    smiles_1, smiles_2, label = row
    # 生成 drugbank_id 和对应的 cid
    drugbank_id_1 = f"DB{drugbank_id_counter:05d}"
    cid_1 = f"CID{cid_counter:08d}"
    if smiles_1 not in drugbank_id_dict:
        drugbank_id_dict[smiles_1] = drugbank_id_1
        cid_dict[smiles_1] = cid_1
        drugbank_id_counter += 1
        cid_counter += 1
    else:
        drugbank_id_1 = drugbank_id_dict[smiles_1]
        cid_1 = cid_dict[smiles_1]
    drugbank_id_2 = f"DB{drugbank_id_counter:05d}"
    cid_2 = f"CID{cid_counter:08d}"
    if smiles_2 not in drugbank_id_dict:
        drugbank_id_dict[smiles_2] = drugbank_id_2
        cid_dict[smiles_2] = cid_2
        drugbank_id_counter += 1
        cid_counter += 1
    else:
        drugbank_id_2 = drugbank_id_dict[smiles_2]
        cid_2 = cid_dict[smiles_2]

    # 将新生成的数据添加到输出列表中
    output_data.append([drugbank_id_1, drugbank_id_2, smiles_2, smiles_1, cid_1, cid_2, label])

# 将数据写入新的 CSV 文件
with open('all_data_deepddi.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['drugbank_id_1', 'drugbank_id_2', 'smiles_2', 'smiles_1', 'cid_1', 'cid_2', 'label'])
    writer.writerows(output_data)

print("生成完成，已保存到all_data_deepddi.csv 文件中。")
```

```python
import pandas as pd

# Step 1: Read drug_list_deepddi.csv and extract required columns
# drug_list = pd.read_csv('drug_list_deepddi.csv')
# drug_list = pd.read_csv('drug_list_chemddi.csv')
drug_list = pd.read_csv('drug_list_hebing.csv')
smiles_cid_map = drug_list.set_index('smiles')[['CID', 'drugbank_id']].to_dict(orient='index')

# Step 2: Define function to update CID and drugbank_id
def update_row(row):
    
    smiles_1 = row['smiles_1']
    smiles_2 = row['smiles_2']
    
    if smiles_1 in smiles_cid_map:
        cid, drugbank_id = smiles_cid_map[smiles_1]['CID'], smiles_cid_map[smiles_1]['drugbank_id']
        row['cid_1'] = cid
        row['drugbank_id_1'] = drugbank_id
    else:
        row['cid_1'] = pd.nan
        row['drugbank_id_1'] = pd.nan
        
    if smiles_2 in smiles_cid_map:
        cid, drugbank_id = smiles_cid_map[smiles_2]['CID'], smiles_cid_map[smiles_2]['drugbank_id']
        row['cid_2'] = cid
        row['drugbank_id_2'] = drugbank_id
    else:
        row['cid_2'] = pd.nan
        row['drugbank_id_2'] = pd.nan
        
    return row

# Step 3: Update files
# file_paths = [
#     './new_data/trainsductive/DEEPDDI/ZhangDDI_train.csv',
#     './new_data/trainsductive/DEEPDDI/ZhangDDI_test.csv',
#     './new_data/trainsductive/DEEPDDI/ZhangDDI_valid.csv',
#     './new_data/type1/DEEPDDI/ZhangDDI_train.csv',
#     './new_data/type1/DEEPDDI/ZhangDDI_test.csv',
#     './new_data/type1/DEEPDDI/ZhangDDI_valid.csv',
#     './new_data/type2/DEEPDDI/ZhangDDI_train.csv',
#     './new_data/type2/DEEPDDI/ZhangDDI_test.csv',
#     './new_data/type2/DEEPDDI/ZhangDDI_valid.csv',
#     'all_data_deepddi.csv'
# ]
# file_paths = [
#     './new_data/trainsductive/chemDDI/ZhangDDI_train.csv',
#     './new_data/trainsductive/chemDDI/ZhangDDI_test.csv',
#     './new_data/trainsductive/chemDDI/ZhangDDI_valid.csv',
#     './new_data/type1/chemDDI/ZhangDDI_train.csv',
#     './new_data/type1/chemDDI/ZhangDDI_test.csv',
#     './new_data/type1/chemDDI/ZhangDDI_valid.csv',
#     './new_data/type2/chemDDI/ZhangDDI_train.csv',
#     './new_data/type2/chemDDI/ZhangDDI_test.csv',
#     './new_data/type2/chemDDI/ZhangDDI_valid.csv',
#     'all_data_chemddi.csv'
# ]
file_paths = [
    './new_data/trainsductive/hebing/ZhangDDI_train.csv',
    './new_data/trainsductive/chemDDI/ZhangDDI_test.csv',
    './new_data/trainsductive/chemDDI/ZhangDDI_valid.csv',
    './new_data/type1/chemDDI/ZhangDDI_train.csv',
    './new_data/type1/chemDDI/ZhangDDI_test.csv',
    './new_data/type1/chemDDI/ZhangDDI_valid.csv',
    './new_data/type2/chemDDI/ZhangDDI_train.csv',
    './new_data/type2/chemDDI/ZhangDDI_test.csv',
    './new_data/type2/chemDDI/ZhangDDI_valid.csv',
    'all_data_chemddi.csv'
]

for file_path in file_paths:
    # Read the file
    df = pd.read_csv(file_path)
    # Update CID and drugbank_id columns
    df = df.apply(update_row, axis=1)
    # Save updated file
    df.to_csv(file_path, index=False)
```

```python
import pandas as pd
import numpy as np

# 定义文件路径列表
file_paths = [
    './new_data/trainsductive/DEEPDDI/ZhangDDI_train.csv',
    './new_data/trainsductive/DEEPDDI/ZhangDDI_test.csv',
    './new_data/trainsductive/DEEPDDI/ZhangDDI_valid.csv',
    './new_data/trainsductive/ZHANGddi/ZhangDDI_train.csv',
    './new_data/trainsductive/ZHANGddi/ZhangDDI_test.csv',
    './new_data/trainsductive/ZHANGddi/ZhangDDI_valid.csv',
    './new_data/trainsductive/chemDDI/ZhangDDI_train.csv',
    './new_data/trainsductive/chemDDI/ZhangDDI_test.csv',
    './new_data/trainsductive/chemDDI/ZhangDDI_valid.csv'
]

# 创建一个空的 DataFrame 用于存储合并后的数据
combined_df = pd.DataFrame()

# 循环读取每个文件并合并到 combined_df 中
for file_path in file_paths:
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# 去除重复行
combined_df.drop_duplicates(inplace=True)

# 保存合并前的数据集为 CSV 文件
combined_df.to_csv('./new_data/trainsductive/hebing/ZhangDDI.csv', index=False)

# 划分数据集为 train、test 和 valid
train_size = int(0.6 * len(combined_df))
test_size = int(0.2 * len(combined_df))
valid_size = len(combined_df) - train_size - test_size

train_df, test_valid_df = np.split(combined_df.sample(frac=1), [train_size])
test_df, valid_df = np.split(test_valid_df, [test_size])

# 保存划分后的数据集为 CSV 文件
train_df.to_csv('./new_data/trainsductive/hebing/ZhangDDI_train.csv', index=False)
test_df.to_csv('./new_data/trainsductive/hebing/ZhangDDI_test.csv', index=False)
valid_df.to_csv('./new_data/trainsductive/hebing/ZhangDDI_valid.csv', index=False)
```
